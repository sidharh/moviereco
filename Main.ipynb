Feature Selection Part

```python
movies=movies[['Movie_id', 'title', 'Genres','Keywords',
       'overview' ]] 
```

```python
movies.columns
```

```python
movies['tags'] = movies['overview'] + movies['Genres']
```

```python
movies.columns
```

```python
new_data = movies.drop(columns=['overview','Genres','Keywords'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector[:10]
```

```python
vector.shape
```

```python
feature_names = cv.get_feature_names_out()
```

```python
feature_names
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"]
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
from operator import itemgetter
distance = sorted(list(enumerate(similarity[2])), key=itemgetter(1), reverse=True)
```

```python
distance[0:5]
```

```python
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), key=itemgetter(1), reverse=True)
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```

```python
recommend("Guardians of the Galaxy Vol. 2")
```

```python
recommend("Guardians of the Galaxy Vol. 2")
```
